```python
from rbnbr.solver.quantum.qrr import QRRnGWMaxCutSolver, QRRMaxCutSolver
```

```python
from rbnbr.problems.max_cut import MaxCutProblem

problem_rnd = MaxCutProblem.generate_random_maxcut_problem(20, 0.5, weighted=False, solve=True)
```

```python
qrr_solver = QRRMaxCutSolver()

qrr_sol = qrr_solver.solve(problem_rnd)
```

```python
qrr_sol.approx_ratio
```

```python


qrrGW_solver = QRRnGWMaxCutSolver()
qrrGW_sol = qrrGW_solver.solve(problem_rnd)
qrrGW_sol.approx_ratio
```

```python
from rbnbr.solver.classical.gw import GWMaxcutSolver
gw_solver = GWMaxcutSolver()
approx_ratios = []
for i in range(100):
    gw_sol = gw_solver.solve(problem_rnd)
    gw_sol.approx_ratio
    approx_ratios.append(gw_sol.approx_ratio)
```

```python

import matplotlib.pyplot as plt
import numpy as np

plt.hist(approx_ratios, bins=20, edgecolor='black')
plt.title(f'{np.mean(approx_ratios)}')
plt.xlabel('Approximation Ratio')
plt.ylabel('Frequency')
plt.show()
```
