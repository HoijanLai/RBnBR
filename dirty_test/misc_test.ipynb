```python
import pickle
with open('target_problem.pkl', 'rb') as f:
    problem_target = pickle.load(f)

problem_info = {
    'problem': problem_target,
    'index_map': list(range(problem_target.N)),
    'compensation': 0
}
```

```python
from rbnbr.solver.BnB.bnb import get_subproblem, BUF
```

```python


root_UF = BUF(problem_target.N)
problem_target.display(use_title=False, show_node_labels=False)
```

```python
L_org = nx.laplacian_matrix(problem_target.graph).toarray()
```

```python
# Create a copy of L_org
L_reduced = L_org.copy()

# Zero out row and column 3
L_reduced[3, :] = 0
L_reduced[:, 3] = 0

# Zero out row and column 7

# Remove zero rows and columns
non_zero_indices = ~np.all(L_reduced == 0, axis=1) & ~np.all(L_reduced == 0, axis=0)
L_reduced = L_reduced[non_zero_indices][:, non_zero_indices]

# Print the shape of the reduced matrix
print(f"Original shape: {L_org.shape}, Reduced shape: {L_reduced.shape}")

eigenvalues = np.linalg.eigvals(L_reduced)
np.max(eigenvalues) * L_reduced.shape[0] / 4


```

```python
from rbnbr.solver.BnB.bnb import get_new_sub_L
root_UF.union(3, 7, 1)
new_L = get_new_sub_L(L_org, 3, 7, root_UF, -1)

# Remove all zero rows and columns from new_L
non_zero_rows = ~np.all(new_L == 0, axis=1)
non_zero_cols = ~np.all(new_L == 0, axis=0)
new_L = new_L[non_zero_rows][:, non_zero_cols]
```

```python

eigenvalues = np.linalg.eigvals(new_L)
np.max(eigenvalues) * new_L.shape[0] / 4
```

```python
root_UF.union(8, 0, 1)
new_L2 = get_new_sub_L(new_L, 8, 0, root_UF, -1)

# Remove all zero rows and columns from new_L
non_zero_rows = ~np.all(new_L2 == 0, axis=1)
non_zero_cols = ~np.all(new_L2 == 0, axis=0)
new_L2 = new_L2[non_zero_rows][:, non_zero_cols]

eigenvalues = np.linalg.eigvals(new_L2)
np.max(eigenvalues) * new_L2.shape[0] / 4

```

```python
root_UF.union(3, 6, 1)
new_L3 = get_new_sub_L(new_L2, 3, 6, root_UF, -1)

# Remove all zero rows and columns from new_L
non_zero_rows = ~np.all(new_L3 == 0, axis=1)
non_zero_cols = ~np.all(new_L3 == 0, axis=0)
new_L3 = new_L3[non_zero_rows][:, non_zero_cols]

eigenvalues = np.linalg.eigvals(new_L3)
np.max(eigenvalues) * (new_L3.shape[0]-3) / 4

```

```python
root_UF.union(8, 0, 1)
new_L4 = get_new_sub_L(new_L3, 1, 3, root_UF, 1)

eigenvalues = np.linalg.eigvals(new_L4)
np.max(eigenvalues) * new_L4.shape[0] / 4
```
